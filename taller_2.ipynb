```python
import pandas as pd
```

```python
url = "/content/diabetes.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Renombrar las columnas
df = df.rename(columns={
    'Pregnancies': 'Embarazos',
    'Glucose': 'Glucosa',
    'BloodPressure': 'Presion_Arterial',
    'SkinThickness': 'Grosor_Piel',
    'Insulin': 'Insulina',
    'BMI': 'IMC',
    'DiabetesPedigreeFunction': 'Funcion_Pedigree_Diabetes',
    'Age': 'Edad',
    'Outcome': 'Resultado'
})

df.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.ensemble import VotingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
import time
```

```python
# Definir X & y
data = df
#X = data.drop('Resultado', axis=1)
#y = data['Resultado']

X = data.iloc[:, : -1]
y = data['Resultado']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Crear instancia de la tecnica KNN
knn = KNeighborsClassifier(n_neighbors=18)

# Tiempo de inicio
tiempo_inicio = time.time()

# Entrenar el modelo
knn.fit(X_train, y_train)

# Realizar predicciones
y_pred = knn.predict(X_test)

# Tiempo después de la predicción
tiempo_final = time.time()

# Calcular el tiempo transcurrido
tiempo_total_entrenamiento = tiempo_final - tiempo_inicio

cm = confusion_matrix(y_test, y_pred)
print(f"cm: {cm}")

precision = accuracy_score(y_test, y_pred)
print(f"precisión: {precision}")

print(f"Tiempo de entrenamiento: {tiempo_total_entrenamiento:.6f} segundos de Knn")
```

```python
# Crear instancia de la tecnica MLP
mlp = MLPClassifier(hidden_layer_sizes=(4,4,4), activation='relu', max_iter=1000, random_state=42)

# Tiempo de inicio
tiempo_inicio = time.time()

# Entrenar el modelo
mlp.fit(X_train, y_train)

# Realizar predicciones
y_pred = mlp.predict(X_test)

# Tiempo después de la predicción
tiempo_final = time.time()

# Calcular el tiempo transcurrido
tiempo_total_entrenamiento = tiempo_final - tiempo_inicio

cm = confusion_matrix(y_test, y_pred)
print(f"cm: {cm}")

accuracy = accuracy_score(y_test, y_pred)
print(f"precisión: {accuracy}")

print(f"Tiempo de entrenamiento: {tiempo_total_entrenamiento:.6f} segundos de MPL")
```

```python
# Crear instancia de Regresion lineal
lr = LinearRegression()

# Tiempo de inicio
tiempo_inicio = time.time()

# Entrenar el modelo
lr.fit(X_train, y_train)

# Realizar predicciones
y_pred = lr.predict(X_test)

# Tiempo después de la predicción
tiempo_final = time.time()

# Calcular el tiempo transcurrido
tiempo_total_entrenamiento = tiempo_final - tiempo_inicio

umbrall = 0.5
y_pred = (y_pred >= umbrall).astype(int)

y_pred

# Calcular la matriz de confusión
cm = confusion_matrix(y_test, y_pred)
print(f"cm: {cm}")

# Calcular la precisión
accuracy = accuracy_score(y_test, y_pred)
print(f"precisión: {accuracy}")

print(f"Tiempo de entrenamiento: {tiempo_total_entrenamiento:.6f} segundos de Regresion Lineal")
```
